# Gradiente de gravidade de uma esfera

## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sphere_grav
```

## Gerando os parâmetros do sistema de coordenadas 

```python
Nx = 100
Ny = 50
area = [-1000.,1000.,-1000.,1000.]
shape = (Nx,Ny)
```

```python
x = np.linspace(area[0],area[1],num=Nx)
y = np.linspace(area[2],area[3],num=Ny) 
yc,xc = np.meshgrid(y,x)
```

```python
voo = -100.
zc = voo*np.ones_like(xc)
```

```python
coordenadas = np.array([yc.ravel(),xc.ravel(),zc.ravel()])
```

## Gerando os parâmetros do prisma

```python
rho = np.array([3000])
modelo = np.array([[0,0,150,100]])
```

## Cálculo das componentes do campo de gravidade e do potencial

```python
gxx = sphere_grav.gravity_gradient(coordenadas,modelo,rho,field="g_xx")
```

```python
gxy = sphere_grav.gravity_gradient(coordenadas,modelo,rho,field="g_xy")
```

```python
gxz = sphere_grav.gravity_gradient(coordenadas,modelo,rho,field="g_xz")
```

```python
gyy = sphere_grav.gravity_gradient(coordenadas,modelo,rho,field="g_yy")
```

```python
gyz = sphere_grav.gravity_gradient(coordenadas,modelo,rho,field="g_yz")
```

```python
gzz = sphere_grav.gravity_gradient(coordenadas,modelo,rho,field="g_zz")
```

```python
laplacian = gxx + gyy + gzz
```

```python
laplacian
```
