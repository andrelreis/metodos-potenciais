# Efeito gravitacional de um poligono

**[Referência bibliográfica]**

* Talwani, M., J. L. Worzel, and M. Landisman (1959), Rapid Gravity Computations for Two-Dimensional Bodies with Application to the Mendocino Submarine Fracture Zone, J. Geophys. Res., 64(1), 49-59, doi:10.1029/JZ064i001p00049

## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import talwani as tw
```

## Calculando o efeito gravitacional de um poligono

```python
xs = np.array([50,150,250,350,450,250,50])
zs = np.array([100,50,75,50,100,200,100])
```

### Visualização da bacia

```python
plt.figure(figsize=(10,10))

plt.xlabel('northing (m)', fontsize = 16)
plt.ylabel('depth (m)', fontsize = 16)
plt.fill(xs,zs,color='red')
plt.xlim(0,500)
plt.ylim(300,0)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

file_name = 'images/polygon_edges'
plt.savefig(file_name+'.png',dpi=300)

plt.show()
```

## Gerando as coordenadas de observação 

```python
xp = np.arange(0,500,step=5)
```

```python
xp.size
```

```python
zp = -100.*np.ones_like(xp)
```

## Calculando o efeito gravitacional ($g_z$) do poligono

```python
gz = tw.gz(xp,zp,xs,zs,density=3000.)
```

## Visualização dos dados observados

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True,figsize=(8,15))

## Efeito gravitacional da bacia
ax1.set_title('Polygon modelling')
ax1.set_ylabel('$g_z$ (mGal)', fontsize = 16)
ax1.plot(xp,gz,'k.')

## Bacia 2D
plt.xlabel('northing (m)', fontsize = 16)
plt.ylabel('depth (m)', fontsize = 16)
plt.fill(xs,zs,color='red')
plt.xlim(0,500)
plt.ylim(300,0)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

file_name = 'images/grav_effect_polygon'
plt.savefig(file_name+'.png',dpi=300)

plt.show()
```
