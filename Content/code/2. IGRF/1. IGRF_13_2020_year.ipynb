## Importando bibliotecas

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cartopy.crs as ccrs
```

```python
notebook_name = '1.IGRF_13_2020_year.ipynb'
```

### Plot style

```python
plt.style.use('ggplot')
```

## Carregando a inclinação

```python
filename = 'data_set/igrf_grid_inclination.csv'
```

```python
names = ['year','latitude','longitude','inclination']
```

```python
data = pd.read_csv(filename,usecols=(0,1,2,4),names=names,skipinitialspace=True,skiprows=13)
```

## Declinação

```python
filename = 'data_set/igrf_grid_declination.csv'
```

```python
names = ['year','latitude','longitude','declination']
```

```python
declination = pd.read_csv(filename,usecols=(0,1,2,4),names=names,skipinitialspace=True,skiprows=13)
```

```python
data['declination'] = declination['declination']
```

## Campo total

```python
filename = 'data_set/igrf_grid_intensity.csv'
```

```python
names = ['year','latitude','longitude','total']
```

```python
intensity = pd.read_csv(filename,usecols=(0,1,2,4),names=names,skipinitialspace=True,skiprows=13)
```

```python
data['intensity'] = intensity['total']
```

```python
data['longitude'].min()
```

## Região da observação

```python
region = (-180,180,-90,90)
```

## Campo para o ano de 2020 (IGRF-13)

```python
title_font = 20
bottom_font = 18
fig = plt.figure(figsize=(10,10),tight_layout=True)
projection = ccrs.Mollweide()

ax1 = plt.axes(projection=projection)
ax1.stock_img()
ax1.set_global()
ax1.set_title('Inclination (Degrees)', size=title_font)
pg1 = ax1.tricontour(data.longitude,
                     data.latitude,
                     data.inclination,50,transform=ccrs.PlateCarree(),colors='k',linewidths=1)
plt.clabel(pg1, inline=2, fontsize=10)

file_name = 'images/inclination_2020'
plt.savefig(file_name+'.png',dpi=300)
plt.show()
```

```python
title_font = 20
bottom_font = 18
fig = plt.figure(figsize=(10,10),tight_layout=True)
projection = ccrs.Mollweide()

ax2 = plt.axes(projection=projection)
ax2.stock_img()
ax2.set_global()
ax2.set_title('Declination (Degrees)', size=title_font)
pg2 = ax2.tricontour(data.longitude,
                     data.latitude,
                     data.declination,70,transform=ccrs.PlateCarree(),colors='k',linewidths=1)
plt.clabel(pg2, inline=2, fontsize=10)

file_name = 'images/declination_2020'
plt.savefig(file_name+'.png',dpi=300)
plt.show()
```

```python
title_font = 20
bottom_font = 18
fig = plt.figure(figsize=(10,10),tight_layout=True)
projection = ccrs.Mollweide()

ax3 = plt.axes(projection=projection)
ax3.stock_img()
ax3.set_global()
ax3.set_title('Total Field (nT)', size=title_font)
pg3 = ax3.tricontour(data.longitude,
                     data.latitude,
                     data.intensity,30,transform=ccrs.PlateCarree(),colors='k',linewidths=1)
plt.clabel(pg3, inline=2, fontsize=10)

file_name = 'images/igrf_2020'
plt.savefig(file_name+'.png',dpi=300)
plt.show()
```
