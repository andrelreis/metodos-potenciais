## Carregando os dados do Complexo de Montes Claros de Goiás

```python
filename = 'data_set/Montes_Claros_data_XYZ_area'
```

```python
names = ['x','y','geometric','microlevel','field','total field','longitude','latitude']
```

```python
data = pd.read_csv(filename,usecols=(0,1,2,3,4,5,6,7),names=names,
                   skipinitialspace=True,delim_whitespace=True).dropna()
```

```python
longitude = data['longitude'].values
latitude = data['latitude'].values
tf_anomaly = data['field'].values
h_geometric = data['geometric'].values
```

## Anomalia de Campo Total - Complexo de Montes Claros

```python
title_font = 20
bottom_font = 18
plt.figure(figsize=(10,10), tight_layout=True)

plt.title('TFA (Montes Claros Complex) ', fontsize=title_font)
plt.xlabel('Longitude (degrees)',fontsize = title_font)
plt.ylabel('Latitude (degrees)',fontsize = title_font)
plt.tricontourf(longitude,latitude,tf_anomaly,30, cmap='jet')
cb = plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
cb.set_label('nT',size=bottom_font)
cb.ax.tick_params(labelsize=bottom_font)

file_name = 'images/montes_claros_real_data'
plt.savefig(file_name+'.png',dpi=300)

plt.show()
```
