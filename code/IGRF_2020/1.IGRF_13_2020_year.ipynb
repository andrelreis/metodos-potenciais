# 13th Generation of the International Geomagnetic Field Reference (IGRF-13)

## Importing libraries

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cartopy.crs as ccrs
import ipywidgets as widgets
```

```python
notebook_name = '1.IGRF_13_2020_year.ipynb'
```

### Plot style

```python
plt.style.use('ggplot')
```

#### Open a dictionary

```python
data_set = dict()
```

#### Saving files

```python
saved_files = []
```

### Loading data

```python
filename = 'data_set/igrf_grid_declination.csv'
```

```python
names = ['YEAR','LAT','LONG','DEC']
```

```python
data = pd.read_csv(filename,usecols=(0,1,2,4),names=names,skipinitialspace=True,skiprows=13).to_numpy()
```

```python
data_set['long'] = data[:,2]
data_set['lat'] = data[:,1]
data_set['dec_world_2020'] = data[:,3]
```

## For the year 2020 (IGRF-13)

```python
title_font = 22
bottom_font = 18
plt.figure(figsize=(15,15), tight_layout=True)

projection = ccrs.Mollweide()
ax = plt.axes(projection=projection)
ax.stock_img()
ax.set_global()
ax.set_title('Declination (Degrees)',size=title_font)
pg = ax.tricontour(data_set['long'],
                    data_set['lat'],
                    data_set['dec_world_2020'],100,transform=ccrs.PlateCarree(),colors='k',linewidths=1)
plt.clabel(pg, inline=2, fontsize=10)

file_name = 'images/dec_world_2020'
plt.savefig(file_name+'.png',dpi=300)
saved_files.append(file_name+'.png')
plt.show()
```
